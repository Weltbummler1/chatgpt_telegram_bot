```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

smtp_server = "smtp.gmail.com"
smtp_port = 587
email_user = "your-email@gmail.com"
email_password = "your-email-password"

# Create the email
msg = MIMEMultipart()
msg["From"] = email_user
msg["To"] = "recipient@example.com"
msg["Subject"] = "Subject of the email"
msg.attach(MIMEText("This is the body of the email.", "plain"))

# Send the email
with smtplib.SMTP(smtp_server, smtp_port) as server:
    server.starttls()  # Required for Gmail's security settings
    server.login(email_user, email_password)
    text = msg.as_string()
    server.sendmail(email_user, "recipient@example.com", text)
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import pymongo

# Set up MongoDB client
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["chatgpt_telegram_bot"]
collection = db["dialog"]

# Email setup
smtp_server = "localhost"
smtp_port = 1025
email_user = "your-email@example.com"

# Fetch conversations by user
users = collection.distinct("user_id")  # get unique user IDs

for user in users:
    conversations = collection.find({"user_id": user})

    # Build conversation text
    conversation_text = ""
    for conversation in conversations:
        conversation_text += f"Conversation started on: {conversation['start_time']['$date']}\n"
        for message in conversation['messages']:
            conversation_text += f"User: {message['user']}\n"
            conversation_text += f"Bot: {message['bot']}\n"
            conversation_text += f"Date: {message['date']['$date']}\n\n"

    # Prepare email
    msg = MIMEMultipart()
    msg["From"] = email_user
    msg["To"] = "recipient@example.com"
    msg["Subject"] = f"Conversations for user {user}"
    msg.attach(MIMEText(conversation_text, "plain"))

    # Send email
    with smtplib.SMTP(smtp_server, smtp_port) as server:
        server.login(email_user, email_password)
        text = msg.as_string()
        server.sendmail(email_user, "recipient@example.com", text)
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import pymongo
from datetime import datetime

# MongoDB connection
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["chatgpt_telegram_bot"]
collection = db["dialog"]

# Fetch conversations from MongoDB
conversations = collection.find()

# Create email content
conversation_text = ""
for conversation in conversations:
    start_time = conversation['start_time']
    start_time = start_time.strftime("%Y-%m-%d %H:%M:%S")
    conversation_text += f"Conversation started on: {start_time}\n"
    for message in conversation['messages']:
        user = message['user']
        bot = message['bot']
        date = message['date']
        date = date.strftime("%Y-%m-%d %H:%M:%S")
        conversation_text += f"User ({date}): {user}\n"
        conversation_text += f"Bot  ({date}): {bot}\n"
    conversation_text += "\n"

# Email configuration
sender_email = "your_sender_email@gmail.com"
receiver_email = "your_receiver_email@gmail.com"
password = "your_email_password"

# Create email message
msg = MIMEMultipart()
msg['From'] = sender_email
msg['To'] = receiver_email
msg['Subject'] = "Conversations from MongoDB"

# Attach email body
msg.attach(MIMEText(conversation_text, 'plain'))

# Send email
with smtplib.SMTP('localhost', 1025) as server:
    server.sendmail(sender_email, receiver_email, msg.as_string())

print("Email sent successfully!")
```

```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import pymongo
import base64

# SMTP server configuration
smtp_host = 'smtp.gmail.com'
smtp_port = 587

# Email configuration
sender_email = 'pilotblogger73@gmail.com'
receiver_email = 'pilotblogger73@gmail.com'
subject = 'AUTOMATED TEST!: Conversations from MongoDB'
password = 'ewmwyghshpvdhobk'

# Set up MongoDB client
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["chatgpt_telegram_bot"]
conversation_collection = db["dialog"]
user_collection = db["user"]


# Retrieve conversations from the collection
conversations = conversation_collection.find()

# Create the email message
message = MIMEMultipart()
message['From'] = sender_email
message['To'] = receiver_email
message['Subject'] = subject

# Plain text content of the email
email_content = ''

# Generate email content from conversations
for conversation in conversations:
    user_id = conversation['user_id']
    start_time = conversation['start_time']
    start_time = start_time.strftime("%Y-%m-%d %H:%M:%S")
    

    user = user_collection.find_one({"_id": user_id})
    user_name = user['username'] if user else 'Unknown Username'
    first_name = user['first_name'] if user else ''
    last_name = user['last_name'] if user else ''

    messages = conversation.get("messages", [])
    if not messages:
        continue  # Skip conversation if no messages

    email_content += f"<b>Conversation started on: {start_time}</b><br><br>"

    email_content += f"User ID: {user_id}<br>"
    email_content += f"<b>User Name: {user_name}</b><br>"
    email_content += f"<b>First Name: {first_name}</b><br>"
    email_content += f"Last Name: {last_name}<br><br>"

    



    for msg in conversation["messages"]:
        user = msg["user"]
        bot = msg["bot"]
        email_content += f"<b>User:</b> {user}<br>"
        email_content += f"<b>Bot:</b> {bot}<br>"

    email_content += "----------------------------------------<br><br>"

# Attach plain text content to the email
message.attach(MIMEText(email_content, 'html'))

# Convert the email message to a string
email_message = message.as_string()

# Send email
with smtplib.SMTP(smtp_host, smtp_port) as server:
    server.starttls()
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, email_message)

# Close the MongoDB connection
client.close()
```
